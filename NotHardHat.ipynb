```python
from ultralytics import YOLO
```

```python
model = YOLO("yolo11n.pt")
```

```python
import yaml
dict_file = {'train':'/content/hardhat-vest-dataset-v3/images/test' ,
            'val': '/content/hardhat-vest-dataset-v3/images/val',
            'names' :
             {0: 'helmet',
             1: 'vest',
             2: 'head',
             3: 'person'}}

with open('/content/hardhat-vest-dataset-v3/hard_head.yaml', 'w') as file:
    documents = yaml.dump(dict_file,file,default_flow_style=False)
```

```python
model.train(data="/content/hardhat-vest-dataset-v3/hard_head.yaml", epochs=20)
```

```python
results = model(["/content/1.jpg"])

for result in results:
    boxes = result.boxes
    masks = result.masks
    keypoints = result.keypoints
    probs = result.probs
    obb = result.obb
    result.show()
    result.save(filename="/content/hardhat-vest-dataset-v3/images/result1.jpg")
```
